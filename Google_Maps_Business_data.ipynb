## Import Required Libraries

```python
import pandas as pd
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
```

## Setting Up Chrome Browser

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach",True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
driver.get('https://www.google.com/search?sca_esv=4920aa5470d17e7b&rlz=1C1CHBD_enIN1020IN1020&tbs=lf:1,lf_ui:2&tbm=lcl&sxsrf=ADLYWIITbzwaD_ADRLSr_bH7TvBQdtLzAQ:1715404791079&q=interior+designers+in+coimbatore&rflfq=1&num=10&sa=X&ved=2ahUKEwiCgt_p7ISGAxUd6jgGHQicDqEQjGp6BAghEAE&biw=1366&bih=633&dpr=1#rlfi=hd:;si:;mv:[[11.099606999999999,77.04963579999999],[10.974595599999999,76.9213712]]')

driver.maximize_window()
sleep(3)
```

## Data Collecting from Google Maps

```python
company = []
add = []
phones = []
web = []
rates = []
review = []
for page in range(1,7):
    #link = driver.find_elements(By.XPATH,"//a[@class='vwVdIc wzN8Ac rllt__link a-no-hover-decoration']")
    for lin in driver.find_elements(By.XPATH,"//a[@class='vwVdIc wzN8Ac rllt__link a-no-hover-decoration']"):
        lin.click()
        sleep(4)
        try:
            company_name = driver.find_element(By.XPATH,"//h2/span")
            company.append(company_name.text)
        except:
            company.append(" ")
        try:
            address = driver.find_element(By.XPATH,"//div/span[@class='LrzXr']")
            add.append(address.text)
        except:
            add.append(address.text)
        try:
            Phone = driver.find_element(By.XPATH,"//a[@data-dtype='d3ph']/span")
            phones.append(Phone.text)
        except:
            phones.append(" ")
        try:
            website = driver.find_element(By.XPATH,"//a[@class='n1obkb']").get_attribute("href")
            web.append(website)
        except:
            web.append(" ")
        try:
            rating = driver.find_element(By.XPATH,"(//span[@class='fzTgPe Aq14fc'])[1]")
            rates.append(rating.text)
        except:
            print(" ")
        try:
            reviews = driver.find_element(By.XPATH,"(//span[@class='z5jxId'])[1]")
            review.append(reviews.text)
        except:
            review.append(" ")
        sleep(3)
        
    nxt = driver.find_element(By.XPATH,"//a[@id='pnnext']")
    nxt.click()
    sleep(4)
```

## Create Data frame for data storing in table

```python
data = pd.DataFrame()
data['Company Name'] = company
data['Address'] = add
data['Phone'] = phones
data['Website URL'] = web
data['Ratings'] = rates
data['Reviews'] = review

data 
```

```python
data.to_csv('Business_data.csv')
```
